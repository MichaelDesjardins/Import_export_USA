```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
export=pd.read_excel("DATAFrames/Standard Report - Exports.xlsx")
export
```

```python
#Using the row 1 containing NaN to remove all columns that are unecessary or duplicated
nan_columns = export.loc[1].isna()
export = export.drop(columns=nan_columns[nan_columns].index)
```

```python
export
```

```python
#Replacing all the "Total" values in the row 4 with "All" meaning all USA districts combined
export.loc[4] = export.loc[4].replace('Total', "All")
```

```python
#Replacing all "Total" value in row 5 with the correct value "2024"
export.loc[5] = export.loc[5].replace('Total', 2024)
```

```python
#Dropping the 2 unnecessary row 0 and 6 containing only NaN
export = export.drop(index=[0,6])
```

```python
export
```

```python
#Flipping the DF making all the columns become rows
export = export.T
```

```python
export
```

```python
#Changing the "Unnamed 0" row title to "Count"
export.index = export.index.where(export.index != "Unnamed: 0", "Count")
```

```python
#Taking the first row of the DF to make it the title of all the columns and removing the old column titles
export.columns = export.loc["Count"]
export = export.drop(index="Count")
export = export.reset_index(drop=True)
```

```python
export
```

```python
#Changing dtype of the "Value ($US)" column into int64 instead of object
export["Value ($US)"] = export["Value ($US)"].astype('int64')
```

```python
#making the DF into a descending order as per the "Value ($US)" column
export=export.sort_values("Value ($US)",ascending=False)
```

```python
#Changing the index to make it in ascending order starting at 1
export.reset_index(drop=True, inplace=True)
export.index = export.index + 1
```

```python
export
```

```python
#Displaying the top 10 category of export from US to Canada
export["Value ($US)"].nlargest(10)
export["Value ($US)"].sum()
```

```python
iimport=pd.read_excel("DATAFrames/Standard Report - Imports.xlsx")
```

```python
#USA imports from Canada
iimport
```

```python
#Dropping unnecessary rows in the dataframe
iimport= iimport.drop(index=[0,1,2,3,4,5])
```

```python
iimport
```

```python
#Changing column title to match the export DF
iimport = iimport.rename(columns={"Standard Report - Imports": "Commodity"})
```

```python
#Making the index start from 1 instead of 0
iimport = iimport.reset_index(drop=True)
iimport.index = iimport.index + 1
```

```python
iimport
```

```python
#Merging the 2 DF using the Commodity column
import_export = iimport.merge(export, on="Commodity", how="outer")
```

```python
import_export
```

```python
#Changing 2 column title containing the export and import values
import_export = import_export.rename(columns={"Unnamed: 1": "Imports value in US$"})
import_export = import_export.rename(columns={"Value ($US)": "Exports value in US$"})
```

```python
#Moving the Import value column at the end of the DF
import_export = import_export[[col for col in import_export if col != "Imports value in US$"] + ["Imports value in US$"]]
```

```python
#Changing the dtype of the import value column
import_export["Imports value in US$"] = import_export["Imports value in US$"].astype("int64")
```

```python
#Dropping the "domestic/Foreign" column
import_export = import_export.drop(columns=["Domestic/Foreign"])
```

```python
#replacing the NaN with the appropriate value
import_export["Country"] = import_export["Country"].fillna("Canada")
import_export["District"] = import_export["District"].fillna("All")
import_export["Time"] = import_export["Time"].fillna(2024)
import_export["Exports value in US$"] = import_export["Exports value in US$"].fillna(0)
```

```python
import_export
```

```python
#Changing Export value column dtype to int64
import_export["Exports value in US$"]=import_export["Exports value in US$"].astype("int64")
```

```python
import_export
```

```python
#Reorganizing the index into a list starting at 1
import_export = import_export.reset_index(drop=True)
import_export.index = import_export.index + 1
```

```python
#Finding the total of exports to Canada
total_export=import_export["Exports value in US$"].sum()
total_export
```

```python
#Finding the total of imports from Canada
total_import=import_export["Imports value in US$"].sum()
total_import
```

```python
#Applying the new function to our DF and creating a new column for it called "Category"
import_export["Category"] = import_export["Commodity"].apply(map_category)
```

```python
#Allowing to display the full Dataframe to look at all the values and displaying the DF
pd.set_option("display.max_rows", None)  #Show all rows
pd.set_option("display.max_columns", None)  #Show all columns
pd.set_option("display.width", None)  #No line breaks, show the DataFrame in a single line
pd.set_option("display.max_colwidth", None)  #Show full column width
import_export
```

```python
#Creating a plot with all the import and export values regrouped into the "Category" column 
grouped_df = import_export.groupby("Category")[["Imports value in US$", "Exports value in US$"]].sum()
grouped_df = grouped_df.sort_values("Imports value in US$", ascending=False)

ax = grouped_df.plot(kind="bar", figsize=(10, 8))
plt.title("Imports and Exports by Category")
plt.xlabel("Category")
plt.ylabel("Value in US$")
plt.xticks(rotation=45, ha="right")
ax.legend(["Imports from Canada", "Exports to Canada"])
#Display the plot
plt.tight_layout()
plt.show()
```

As shown in the data, most categories have higher imports from Canada than exports to Canada. Following the 25% tariff on Canadian products, Canada responded by imposing a 25% tariff on $155 billion worth of American goods, representing approximately more then 50% of all products shipped to Canada annually. These tariffs will affect both companies and citizens in both countries. In the short term, Canadian companies and American consumers will suffer the most, facing layoffs, reduced production, and a significant decline in purchasing power for Americans, as Canada exports more to the U.S. than vice versa. However, both countries may increasingly turn to other import sources and focus on strengthening domestic industries, which could benefit them in the long run.
